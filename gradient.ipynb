```python
def model(fi0,M0,alpha,M):
    fi=0.4*np.log(10)*(fi0)*(10**(-0.4*(M-M0)*(alpha+1)))*np.exp(-10**(-0.4*(M-M0)))
    return(fi)
```

```python
Gc2=np.zeros(3)
def GRAD(fi0,M0,alpha,M,L):
    sigma= L * 0.1
    
    dfi0 = model(fi0,M0,alpha,M)/fi0
    dchi2_fi0 = ((model(fi0,M0,alpha,M)-L)*dfi0)/sigma
    dchi2_fi0 = np.sum(dchi2_fi0)
    
    dM0 = model(fi0,M0,alpha,M)*0.4*np.log(10)*((alpha+1)-10**(-0.4*(M-M0)))
    dchi2_M0 = ((model(fi0,M0,alpha,M)-L)*dM0)/sigma
    dchi2_M0 = np.sum(dchi2_M0)
    
    dalpha = model(fi0,M0,alpha,M)*(0.4*(M-M0)*np.log(10))
    dchi2_alpha = ((model(fi0,M0,alpha,M)-L)*fi0)/sigma
    dchi2_alpha = np.sum(dchi2_alpha) 
   
    
    Gc2[0] = dchi2_fi0
    Gc2[1] = dchi2_M0
    Gc2[2] = dchi2_alpha
    
    return (Gc2)
```
